# 📚 COMPLETE GUIDE: Types of Regression & When to Use Them

---

## 🤔 What is Regression?

**Regression** is a supervised learning technique used to **predict continuous numerical values** (not categories).

**Examples of regression problems:**
- Predicting house prices 🏠
- Predicting exam scores 📊
- Predicting temperature 🌡️
- Predicting sales numbers 💰

---

## 📖 Types of Regression (Most Common)

### 1. **Linear Regression** 📏

**Equation:** `y = w*x + b` (for one feature)

**When to use:**
- ✅ Data shows a **straight-line relationship**
- ✅ Simple and interpretable results needed
- ✅ Starting point for any regression analysis

**Example:** 
- Study hours vs Exam scores
- Experience years vs Salary
- Ad spending vs Sales (when relationship is linear)

**Visual Pattern:**
```
    y
    |        •
    |      •
    |    •
    |  •
    |•___________x
    (Straight line!)
```

**Pros:**
- Simple and fast
- Easy to interpret
- Good baseline model

**Cons:**
- Can't capture curved patterns
- Assumes linear relationship

---

### 2. **Polynomial Regression** 📈

**Equation:** `y = w₀ + w₁*x + w₂*x² + w₃*x³ + ...`

**When to use:**
- ✅ Data shows a **curved/non-linear relationship**
- ✅ Linear regression doesn't fit well
- ✅ Relationship has peaks, valleys, or bends

**Example:**
- Temperature throughout the day (goes up then down)
- Speed vs Fuel efficiency (optimal point exists)
- Age vs Income (increases then plateaus)

**Visual Pattern:**
```
    y
    |      •••
    |    •     •
    |  •         •
    | •
    |•_____________x
    (Curved line!)
```

**Degree matters:**
- Degree 2: U-shaped or inverted U
- Degree 3: S-shaped curves
- Higher degrees: More complex curves

**Pros:**
- Captures non-linear patterns
- Flexible

**Cons:**
- Can overfit easily
- Harder to interpret
- More computationally expensive

---

### 3. **Multiple Linear Regression** 📊

**Equation:** `y = w₁*x₁ + w₂*x₂ + w₃*x₃ + ... + b`

**When to use:**
- ✅ You have **multiple features** (predictors)
- ✅ Want to understand impact of each feature
- ✅ Features contribute linearly to target

**Example:**
- House price = f(size, bedrooms, location, age)
- Salary = f(experience, education, skills)
- Sales = f(TV ads, Radio ads, Newspaper ads) ← **This lab!**

**Visual:** (Hard to draw in 2D - it's a plane or hyperplane in multi-dimensional space)

**Pros:**
- Handles multiple inputs
- Shows importance of each feature

**Cons:**
- Assumes linear relationships
- Can have multicollinearity issues

---

### 4. **Ridge Regression (L2 Regularization)** 🛡️

**Equation:** `y = w₁*x₁ + w₂*x₂ + ... + b` (with penalty on large w values)

**When to use:**
- ✅ You have **many features**
- ✅ Features are **correlated** (multicollinearity)
- ✅ Want to prevent overfitting

**Example:**
- Predicting with 100+ features
- Medical data with correlated measurements

**What it does:** Shrinks coefficient values to prevent overfitting

---

### 5. **Lasso Regression (L1 Regularization)** 🎯

**Equation:** Similar to Ridge, but different penalty

**When to use:**
- ✅ You have **many features**
- ✅ Want **automatic feature selection**
- ✅ Suspect some features are useless

**What it does:** Can set some coefficients to exactly 0 (removes features)

---

### 6. **Logistic Regression** ⚠️

**Note:** Despite the name, this is for **CLASSIFICATION**, not regression!

**When to use:**
- ✅ Predicting **categories** (Yes/No, Spam/Not Spam)
- ❌ NOT for continuous values!

**Example:**
- Will customer buy? (Yes/No)
- Is email spam? (Yes/No)

---

## 🎯 DECISION TREE: Which Regression to Use?

```
START
  │
  ├─ Predicting categories? 
  │   └─ YES → Use CLASSIFICATION (Logistic Regression, etc.)
  │   └─ NO ↓
  │
  ├─ How many features?
  │   ├─ One feature
  │   │   ├─ Straight line pattern? → LINEAR REGRESSION
  │   │   └─ Curved pattern? → POLYNOMIAL REGRESSION
  │   │
  │   └─ Multiple features
  │       ├─ Simple model? → MULTIPLE LINEAR REGRESSION
  │       ├─ Many features + correlated? → RIDGE REGRESSION
  │       └─ Want feature selection? → LASSO REGRESSION
```

## 💡 Practical Tips

### 🔍 How to Choose:

1. **Start simple:** Always try Linear Regression first!
2. **Check the plot:** Does data look straight or curved?
3. **Evaluate performance:** Use RMSE, R², etc.
4. **Try next level:** If linear doesn't work, try polynomial or multiple
5. **Avoid overfitting:** Don't use complex models on simple data!

### ⚠️ Common Mistakes:

- ❌ Using polynomial regression when linear would work (overfitting)
- ❌ Using linear when data is clearly curved (underfitting)
- ❌ Not checking if features are correlated in multiple regression
- ❌ Using too high polynomial degree (degree 10+ is usually bad!)

### ✅ Best Practices:

- Always plot your data first!
- Start with simplest model
- Compare multiple models
- Use train/test split to evaluate
- Check residuals (errors) to validate assumptions

````python
# Visual comparison: Linear vs Polynomial Regression
import numpy as np
import matplotlib.pyplot as plt

# Create sample data with a curved pattern
np.random.seed(42)
x_demo = np.linspace(0, 10, 50)
y_demo = 2 + 3*x_demo - 0.5*x_demo**2 + np.random.randn(50)*3  # Quadratic with noise

# Fit different models
from numpy.polynomial import polynomial as P

# Linear fit (degree 1)
linear_coef = np.polyfit(x_demo, y_demo, 1)
y_linear = np.polyval(linear_coef, x_demo)

# Polynomial fit (degree 2)
poly2_coef = np.polyfit(x_demo, y_demo, 2)
y_poly2 = np.polyval(poly2_coef, x_demo)

# Polynomial fit (degree 5) - might overfit
poly5_coef = np.polyfit(x_demo, y_demo, 5)
y_poly5 = np.polyval(poly5_coef, x_demo)

# Create comparison plot
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Linear Regression
axes[0].scatter(x_demo, y_demo, alpha=0.6, s=50, label='Data Points')
axes[0].plot(x_demo, y_linear, 'r-', linewidth=2, label='Linear Fit (Degree 1)')
axes[0].set_title('LINEAR REGRESSION\n(Underfitting - Too Simple)', fontsize=13, fontweight='bold')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].text(0.5, 0.95, '⚠️ Misses curved pattern!', 
             transform=axes[0].transAxes, ha='center', va='top',
             bbox=dict(boxstyle='round', facecolor='orange', alpha=0.7))

# Good Polynomial (Degree 2)
axes[1].scatter(x_demo, y_demo, alpha=0.6, s=50, label='Data Points')
axes[1].plot(x_demo, y_poly2, 'g-', linewidth=2, label='Polynomial Fit (Degree 2)')
axes[1].set_title('POLYNOMIAL REGRESSION (Degree 2)\n(Good Fit! ✅)', fontsize=13, fontweight='bold', color='green')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].text(0.5, 0.95, '✅ Captures pattern well!', 
             transform=axes[1].transAxes, ha='center', va='top',
             bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7))

# Overfit Polynomial (Degree 5)
axes[2].scatter(x_demo, y_demo, alpha=0.6, s=50, label='Data Points')
axes[2].plot(x_demo, y_poly5, 'purple', linewidth=2, label='Polynomial Fit (Degree 5)')
axes[2].set_title('POLYNOMIAL REGRESSION (Degree 5)\n(Overfitting - Too Complex)', fontsize=13, fontweight='bold')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].legend()
axes[2].grid(True, alpha=0.3)
axes[2].text(0.5, 0.95, '⚠️ Too wiggly!', 
             transform=axes[2].transAxes, ha='center', va='top',
             bbox=dict(boxstyle='round', facecolor='orange', alpha=0.7))

plt.suptitle('COMPARING REGRESSION TYPES: Finding the Right Complexity', 
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("="*70)
print("KEY INSIGHTS FROM THE VISUALIZATION:")
print("="*70)
print("📊 LEFT (Linear): Too simple - can't capture the curve")
print("   → UNDERFITTING: Model is too simple for the data")
print()
print("✅ MIDDLE (Polynomial Degree 2): Just right - follows the pattern")
print("   → GOOD FIT: Model complexity matches data complexity")
print()
print("⚠️  RIGHT (Polynomial Degree 5): Too complex - wiggles too much")
print("   → OVERFITTING: Model is too complex, fits noise not signal")
print()
print("💡 LESSON: Choose model complexity based on your data pattern!")
print("="*70)
````

## 🤔 Understanding `np.polyfit()` and `np.polyval()`

### Quick Explanation:

**`np.polyfit()`** - Finds the best coefficients (w values and b)  
**`np.polyval()`** - Uses those coefficients to calculate predictions

Think of it like a 2-step process:
1. **Training** (polyfit): Find the best parameters
2. **Prediction** (polyval): Use those parameters to make predictions

---

### 📚 What is `np.polyval()`?

**`np.polyval(coefficients, x_values)`**

**Purpose:** Evaluates a polynomial at given x values using the coefficients.

**Simple explanation:** 
> It takes the coefficients (w and b) from `np.polyfit()` and calculates the y values for any x you give it.

---

### 🔢 Example: Linear Regression (Degree 1)

```python
# Step 1: Fit the model (TRAINING)
coefficients = np.polyfit(x, y, 1)
# Returns: [w, b]  (slope and intercept)
# Example: [2.5, 10]  means  y = 2.5*x + 10

# Step 2: Make predictions (USING THE MODEL)
y_predictions = np.polyval(coefficients, x)
# This calculates: y = 2.5*x + 10 for each x value
```

**What polyval does:**
- Takes `coefficients = [2.5, 10]`
- For each x value, calculates: `y = 2.5*x + 10`
- Returns array of predictions

---

### 📈 Example: Polynomial Regression (Degree 2)

```python
# Step 1: Fit the model
coefficients = np.polyfit(x, y, 2)
# Returns: [w2, w1, b]  (three coefficients)
# Example: [-0.5, 3, 2]  means  y = -0.5*x² + 3*x + 2

# Step 2: Make predictions
y_predictions = np.polyval(coefficients, x)
# This calculates: y = -0.5*x² + 3*x + 2 for each x
```

**What polyval does:**
- Takes `coefficients = [-0.5, 3, 2]`
- For each x value, calculates: `y = -0.5*x² + 3*x + 2`
- Returns array of predictions

---

### 🎯 Side-by-Side Comparison

| Step | Function | What It Does | Output |
|------|----------|--------------|--------|
| **1. Training** | `np.polyfit(x, y, deg)` | Finds best coefficients from data | `[w2, w1, b]` (array of numbers) |
| **2. Prediction** | `np.polyval(coefs, x)` | Calculates y values using coefficients | `[y1, y2, y3, ...]` (predictions) |

---

### 💻 Complete Example with Explanations

```python
import numpy as np

# Our data
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 7, 13, 21, 31])

# STEP 1: Training - find coefficients
coefficients = np.polyfit(x, y, 2)  # Fit polynomial of degree 2
print(f"Coefficients: {coefficients}")
# Output: [ 1.  1.  1.]
# This means: y = 1*x² + 1*x + 1

# STEP 2: Prediction - use coefficients
y_pred = np.polyval(coefficients, x)
print(f"Predictions: {y_pred}")
# Output: [ 3.  7. 13. 21. 31.]
# These are the predicted y values!

# Test on new data
new_x = 6
new_y_pred = np.polyval(coefficients, new_x)
print(f"Prediction for x=6: {new_y_pred}")
# Output: 43  (because 1*6² + 1*6 + 1 = 43)
```

---

### 🧮 What's Happening Under the Hood?

**For degree 1 (Linear):**
```python
coefficients = [w, b]
np.polyval([w, b], x) 
# Calculates: w*x + b
```

**For degree 2 (Quadratic):**
```python
coefficients = [w2, w1, b]
np.polyval([w2, w1, b], x)
# Calculates: w2*x² + w1*x + b
```

**For degree 3 (Cubic):**
```python
coefficients = [w3, w2, w1, b]
np.polyval([w3, w2, w1, b], x)
# Calculates: w3*x³ + w2*x² + w1*x + b
```

---

### ✅ Key Points to Remember

1. **`np.polyfit()`** = Training (finds the best w and b)
2. **`np.polyval()`** = Prediction (uses w and b to calculate y)
3. **Order matters!** Coefficients are returned from highest power to lowest
   - For degree 2: `[w2, w1, b]` not `[b, w1, w2]`
4. **You always use them together:**
   - First polyfit to train
   - Then polyval to predict

---

### 🆚 Alternative: Manual Calculation

Instead of `np.polyval()`, you could calculate manually:

```python
# Using polyval (EASY!)
y_pred = np.polyval([w2, w1, b], x)

# Same result manually (MORE WORK!)
y_pred = w2*x**2 + w1*x + b
```

**Why use polyval?**
- ✅ Works for any polynomial degree automatically
- ✅ Less error-prone
- ✅ Cleaner code
- ✅ Handles arrays efficiently

---

### 📝 Common Pattern You'll See

```python
# This pattern appears everywhere in polynomial regression:

# 1. Fit (training)
coefs = np.polyfit(x_train, y_train, degree)

# 2. Predict (inference)
y_pred = np.polyval(coefs, x_test)

# 3. Evaluate
error = np.mean((y_test - y_pred)**2)
```

### 🎓 Summary in One Sentence

**`np.polyval()` takes the coefficients from `np.polyfit()` and calculates the predicted y values for any x values you provide.**

````python
# 🎯 DEMONSTRATION: polyfit vs polyval in action

import numpy as np
import matplotlib.pyplot as plt

print("="*70)
print("POLYFIT AND POLYVAL DEMONSTRATION")
print("="*70)

# Create simple data
x_simple = np.array([1, 2, 3, 4, 5])
y_simple = np.array([2, 4, 5, 4, 2])  # Looks like an upside-down U

print("\n📊 Our Data:")
print(f"x: {x_simple}")
print(f"y: {y_simple}")

# STEP 1: Use polyfit to find coefficients (TRAINING)
print("\n" + "="*70)
print("STEP 1: TRAINING with np.polyfit()")
print("="*70)

degree = 2
coefficients = np.polyfit(x_simple, y_simple, degree)

print(f"\nFitting a polynomial of degree {degree}")
print(f"Coefficients found: {coefficients}")
print(f"\nBreaking it down:")
print(f"  w2 (x² coefficient) = {coefficients[0]:.4f}")
print(f"  w1 (x coefficient)  = {coefficients[1]:.4f}")
print(f"  b  (intercept)      = {coefficients[2]:.4f}")
print(f"\n✅ Model equation: y = {coefficients[0]:.4f}*x² + {coefficients[1]:.4f}*x + {coefficients[2]:.4f}")

# STEP 2: Use polyval to make predictions (USING THE MODEL)
print("\n" + "="*70)
print("STEP 2: PREDICTION with np.polyval()")
print("="*70)

y_predictions = np.polyval(coefficients, x_simple)

print(f"\nUsing coefficients to predict y for each x:")
print(f"Predictions: {y_predictions}")

# Compare actual vs predicted
print("\n📊 Comparison: Actual vs Predicted")
print("-"*70)
print(f"{'x':<10} {'Actual y':<15} {'Predicted y':<15} {'Error':<10}")
print("-"*70)
for i in range(len(x_simple)):
    error = abs(y_simple[i] - y_predictions[i])
    print(f"{x_simple[i]:<10} {y_simple[i]:<15.2f} {y_predictions[i]:<15.2f} {error:<10.4f}")

# STEP 3: Test on new data
print("\n" + "="*70)
print("STEP 3: PREDICT ON NEW DATA")
print("="*70)

new_x = 3.5
new_prediction = np.polyval(coefficients, new_x)
print(f"\n❓ What if x = {new_x}?")
print(f"✅ Predicted y = {new_prediction:.4f}")
print(f"\nCalculation breakdown:")
print(f"  y = {coefficients[0]:.4f}*(3.5)² + {coefficients[1]:.4f}*(3.5) + {coefficients[2]:.4f}")
print(f"  y = {coefficients[0]*3.5**2:.4f} + {coefficients[1]*3.5:.4f} + {coefficients[2]:.4f}")
print(f"  y = {new_prediction:.4f}")

# Visual representation
print("\n" + "="*70)
print("VISUAL REPRESENTATION")
print("="*70)

plt.figure(figsize=(10, 6))

# Plot actual data
plt.scatter(x_simple, y_simple, s=100, c='blue', marker='o', 
            label='Actual Data', zorder=3, edgecolors='black', linewidth=2)

# Create smooth line for predictions
x_smooth = np.linspace(0.5, 5.5, 100)
y_smooth = np.polyval(coefficients, x_smooth)
plt.plot(x_smooth, y_smooth, 'r-', linewidth=2, 
         label=f'Fitted Polynomial (Degree {degree})', zorder=2)

# Plot predictions at data points
plt.scatter(x_simple, y_predictions, s=100, c='red', marker='x', 
            label='Predictions (from polyval)', zorder=3, linewidths=3)

# Plot new prediction
plt.scatter(new_x, new_prediction, s=200, c='green', marker='*', 
            label=f'New Prediction (x={new_x})', zorder=4, edgecolors='black', linewidth=2)

# Add error lines
for i in range(len(x_simple)):
    plt.plot([x_simple[i], x_simple[i]], [y_simple[i], y_predictions[i]], 
             'gray', linestyle='--', alpha=0.5, zorder=1)

plt.title('polyfit() finds coefficients, polyval() uses them to predict', 
          fontsize=14, fontweight='bold')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n" + "="*70)
print("✅ SUMMARY:")
print("="*70)
print("1. np.polyfit() analyzed the data and found the best coefficients")
print("2. np.polyval() used those coefficients to calculate predictions")
print("3. We can use polyval() on ANY x value to get predictions!")
print("="*70)
````

coefs = np.polyfit(x, y, 3)      # Returns: [w3, w2, w1, b]

### Common Mistakes ❌ → ✅

**❌ Wrong:**
```python
# Trying to predict without fitting first
y_pred = np.polyval([1, 2], x)  # Where did [1, 2] come from?
```

**✅ Right:**
```python
# Always fit first, then predict
coefs = np.polyfit(x, y, 1)     # Get coefficients from data
y_pred = np.polyval(coefs, x)   # Use those coefficients
```

---

**❌ Wrong:**
```python
# Mixing up the order
y_pred = np.polyfit(coefs, x)   # polyfit is for training!
```

**✅ Right:**
```python
# polyfit = train, polyval = predict
coefs = np.polyfit(x, y, deg)   # Training
y_pred = np.polyval(coefs, x)   # Predicting
```

---

### Memory Trick 🧠

**polyFIT** = **FIT**ting the model (training) → "Find It, Train!"  
**polyVAL** = **VAL**ues/eval**UA**tion (predicting) → "Values, Apply Lost!"

**Or simply:**
- **FIT** first (polyfit)
- **VAL**ue second (polyval)

# ✅ DIRECT ANSWER: What is `np.polyval()`?

## Simple Answer:

**`np.polyval()` calculates predicted y values using polynomial coefficients.**

---

## The Complete Picture:

```
┌─────────────────────────────────────────────────────────────┐
│                    THE WORKFLOW                             │
└─────────────────────────────────────────────────────────────┘

    📊 DATA                    🎓 TRAINING                   🔮 PREDICTION
    
  x = [1,2,3,4,5]         coefficients =              y_pred = 
  y = [3,5,7,9,11]        np.polyfit(x, y, 1)        np.polyval(coefficients, x)
                                 ↓                              ↓
                          Returns: [2, 1]             Uses those to calculate
                          (meaning y=2x+1)            predictions!
                          
                          ┌─────────────┐              ┌─────────────┐
                          │   polyfit   │              │   polyval   │
                          └─────────────┘              └─────────────┘
                          FINDS the best               USES the 
                          coefficients                 coefficients
```

---

## Think of it Like This: 🍳

### Making a Recipe (Training):
```python
recipe = np.polyfit(ingredients, taste_scores, 2)
# "After testing, the best recipe is: 2 cups flour + 1 cup sugar + 3 eggs"
```

### Following the Recipe (Prediction):
```python
result = np.polyval(recipe, new_ingredients)
# "Using that recipe, this new batch will taste like..."
```

---

## The Math Connection:

**For degree 2 polynomial:**

```python
# What polyfit returns:
coefficients = [w2, w1, b]  # Example: [-0.5, 3, 2]

# What polyval calculates:
y = w2*x² + w1*x + b
y = -0.5*x² + 3*x + 2  # For each x value
```

**`polyval()` is just a convenient way to calculate the polynomial equation!**

## Why Use It? 🤷

Instead of writing:
```python

```

You can write:
```python
y_pred = np.polyval([w2, w1, b], x)  # Automatic!
```

**Benefits:**
- ✅ Works for ANY degree (1, 2, 3, 100...)
- ✅ No chance of typos
- ✅ Handles arrays efficiently
- ✅ Cleaner, more readable code

## Bottom Line:

**`polyfit`** = Find the best equation  
**`polyval`** = Use that equation to predict

| What You See | What to Use | Example |
|--------------|-------------|---------|
| **Straight line** 📏 | Linear Regression | `y = w*x + b` |
| **U-shape or inverted U** 📈 | Polynomial (Degree 2) | `y = w₂*x² + w₁*x + b` |
| **S-curve** 〰️ | Polynomial (Degree 3) | `y = w₃*x³ + w₂*x² + w₁*x + b` |
| **Multiple inputs** 📊 | Multiple Linear | `y = w₁*x₁ + w₂*x₂ + ... + b` |
| **Complex wiggly** 🌊 | Try degree 4-5, but be careful! | Risk of overfitting |
| **No clear pattern** ❓ | Linear first, then try others | Start simple |

---

### Step 3: Evaluate and Adjust 📊

**After fitting your model:**

1. **Check RMSE** - Lower is better
2. **Plot predictions vs actual** - Should be close
3. **Look at residuals** - Should be random, not patterned

**If model is bad:**
- Linear didn't work? → Try polynomial
- Polynomial too wiggly? → Lower the degree
- Multiple features confusing? → Try feature selection

---

## 🚦 Traffic Light System for Model Selection

### 🟢 GREEN LIGHT (Safe choices):

✅ **Always start with Linear Regression**
- Fast, simple, interpretable
- Good baseline to compare against

✅ **Use Polynomial Degree 2-3 for curves**
- Handles most real-world curved patterns
- Not too complex

✅ **Multiple Linear for multiple features**
- Standard approach for many predictors

---

### 🟡 YELLOW LIGHT (Use with caution):

⚠️ **Polynomial Degree 4-6**
- Might be needed for complex patterns
- Watch out for overfitting
- Always use train/test split!

⚠️ **Ridge/Lasso with many features**
- Great for high-dimensional data
- Requires tuning hyperparameters

---

### 🔴 RED LIGHT (Usually avoid):

❌ **Polynomial Degree 10+**
- Almost always overfits!
- Too complex for most real data

❌ **Complex model on small dataset**
- Need enough data for complex models
- Rule of thumb: 10+ data points per parameter

## Linear regression

Linear regression is a supervised learning algorithm, the predicted output is continuous. 

When we have one feature (also called predictor) variable it's called simple linear regression. A linear regression model with one feature variable looks like this:

$f_{w,b}(x) = w \cdot x + b$

The output of this model for any given x, is called the prediction of the model for that x. 

We done the output of the model $\hat{y}$. In other words, we have

$\hat{y}^{i} = w \cdot x^{i} + b$,

where $x$ is the predictor (feature) variable, $\hat{y}$ is the prediction, $w$ is the slope and $b$ the intercept of our model. 

The parameters of our model are unknown and needs to be estimated using our training data points $(x^1, y^1), (x^2, y^2), \ldots, (x^m,y^m)$.


**Example**

Let's try to predict sales numbers using only one feature. Let's choose TV spending as the feature, since that seems to have a better linear relationship with the target (sales numbers) than the other features. We draw this conclusion using visual inspection of our scatterplots.

````python
x_train, y_train = df["TV"], df["Sales"]

# fits a polynomial of degree deg using least squares polynomial fit.

w, b = np.polyfit(x_train, y_train, deg=1)  # returns coefficient with highest power first
````

````python
print('Estimated model parameters, based on our training data')
print(f'w : {w}')
print(f'b : {b}')
````

````python
# let's define our linear model

def linear_model(x, w, b):

    y_hat = w*x + b
    
    return y_hat
````

````python
# beräkna nu ut våra predictions, för varje punkt i vår dataset

y_hats = [linear_model(x, w, b) for x in x_train]
````

## Polynomial regression

Let's now try polynomial regression instead. This is just another another regression algorithm which might be better suited to some data than a linear regression model.  

Let's assume we want to fit a polynomial of degree two. If we also have one feature, the polyonomial regression model of degree two then  looks like this:

$f_{\bold{w},b}(x) = w_2 \cdot x^2 + w_1 \cdot x + b$.

The prediction, for training simple $i$ is thus denoted

$\hat{y}^{i} = w_2 \cdot (x^{{i}})^2 + w_1 \cdot x^{i} + b$,

````python
w2, w1, b = np.polyfit(x_train, y_train, deg=2)  # returns coefficient with highest power first

print('Estimated model parameters, based on our training data')
print(f'w2 : {w2}')
print(f'w1 : {w1}')
print(f'b  : {b}')
````

````python
# let's now define our polynomial model (of degree two)

def polynomial_model(x, w2, w1, b):

    y_hat = w2*x**2 + w1*x + b
    
    return y_hat
````

````python
# beräkna våra predicted y-values

y_hats = [polynomial_model(x, w2, w1, b) for x in x_train]
````

````python
# compare the resulting predictions with the ground truth

fig, ax = plt.figure(figsize=(5, 3), dpi=100), plt.axes()

sns.scatterplot(data=df, x="TV", y="Sales")
sns.lineplot(x=x_train, y=y_hats, color="red")

ax.set(
    title="TV advertisement linear regression",
    xlabel="Thousands dollars",
    ylabel="Sales thousands units",
)
````

It indeed, atleast looks, as if this is a better fitting model to our data!

Evaluating the models - how well did we predict $\bf{y}$ with $\hat{\bf{y}}$?

To answer this question we use several **evaluation metrics** or **loss functions**: 

- Mean Absolute Error (MAE) - mean of error between $\bf{y}$ and ${\hat{\bf{y}}}$. The unit is same as measured quantity.

$$MAE = \frac{1}{m}\sum_{i=1}^m |y^i - \hat{y}^i|$$

- Mean Squared Error (MSE) - mean of squared errors between $\bf{y}$ and ${\hat{\bf{y}}}$. It punishes large errors, and the units are in square units of the measured quantity

$$MSE = \frac{1}{m}\sum_{i=1}^m (y^i - \hat{y}^i)^2$$

- Root Mean Squared Error (RMSE) - square root of MSE between $\bf{y}$ and ${\hat{\bf{y}}}$. It punishes large errors, and the units are same as measured quantity, hence easier to interpret.

$$RMSE = \sqrt{\frac{1}{m}\sum_{i=1}^m (y^i - \hat{y}^i)^2}$$
